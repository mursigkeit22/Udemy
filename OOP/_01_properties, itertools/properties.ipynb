### First option:
without underscore in init, underscore inside property.
Important: setter will be called in init
    

```python
# without underscore in init, underscore inside property
class Account:
  
    def __init__(self,  first_name, ):
        
        self.first_name = first_name
   
    @property 
    def first_name(self):
        print("calling getter")
        return self._first_name
    
    @first_name.setter
    def first_name(self, value):
        print("calling setter")
        self._first_name = Account.validate_name(value, 'First Name')
        
  
    
    @staticmethod
    def validate_name(value, field_title):
        if value is None or len(str(value).strip()) == 0:
            raise ValueError(f'{field_title} cannot be empty.')
        return str(value).strip()
    
a = Account('hop')#calling setter
print(a.__dict__) # only '_first_name'
print(a.first_name)#calling getter #hop
print(a._first_name)#hop
a.first_name='pop'#calling setter
print(a.first_name)#calling getter #pop
print(a._first_name)#pop
a._first_name = "third"#nothing
print(a.first_name)#calling getter #third

#works properly
```
